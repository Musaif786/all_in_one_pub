```python
df_m1 = pd.DataFrame({
    "ID": [1,2,3,4],
    "Name": ["Musaif","Noman","Manu","Rajan"]
})
df_m1
```

```python
df_m2 = pd.DataFrame({
    "ID": [1,2,3,5],
    "City": ["Hyd","Bang","Del","Up"]
})
df_m2
```

```python
pd.merge(df_m1,df_m2, on="ID", how="outer")
```

```python
pd.merge(df_m1,df_m2, on="ID", how="left")
```

```python
pd.merge(df_m1,df_m2, how="right" )
```

```python
pd.merge(df_m1,df_m2, how="outer" ,indicator=True )
```

```python
pd.merge(df_m1,df_m2, how="inner" ,indicator=True )
```
