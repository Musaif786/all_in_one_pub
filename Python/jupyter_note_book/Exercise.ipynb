```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
plt.xlabel("Year")
plt.ylabel("per capital")
plt.scatter(df.year, df.percapital,color="red", marker ="+")
```

```python
new_df = df.drop("percapital",axis="columns")
new_df
```

```python
model = linear_model.LinearRegression()
model.fit(new_df,df.percapital)
```

```python
model.predict([[2020]])
```
