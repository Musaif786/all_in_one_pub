```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days = np.arange(1,16)
temp = [36.6,37,37,39,40,41,45,46,47,49,49.5,50,50.1,51,52]
```

```python
plt.title("Random Temperature")
plt.xlabel("Days")
plt.ylabel("Temp")
plt.plot(days,temp, color="g", marker = "o",linestyle="--")
plt.show()
plt.scatter(days,temp)
```

```python
plt.title("Random Temperature")
plt.xlabel("Days")
plt.ylabel("Temp")
plt.plot(days,temp, "ro--")
plt.show()
```

```python
import random

s1 = np.random.randint(18,45,(100))
s2 = np.random.randint(15,40,(100))
```
